```python
import tweepy

CONSUMER_KEY = 'uMoFkw4oKaVs87FmwzskgqhVj'
CONSUMER_SECRET = 'Sfo3DLi7OtHgUEMgzZJGqumSfFb62Hr3IIvPNhaRNReTxCpYpH'

def oauth(consumer_key, consumer_secret):    
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth_url = auth.get_authorization_url()
    
    verify_code = input("Authenticate at %s and then enter you verification code here > " % auth_url) 
    auth.get_access_token(verify_code)
    
    return tweepy.API(auth)

def delete_all(api):
    saveData = []        
    for status in tweepy.Cursor(api.user_timeline).items():
        saveData.append (str(status.created_at) + " | " + str(status.text))
        try:
            api.destroy_status(status.id)
            print ("Deleted: ", status.id)
        except:
            print ("Failed to delete: ", status.id + status.text)

    outF = open("allSafaTweetsRd2.txt", "w")
    for line in saveData:
      outF.write(line)
      outF.write("\n")
    outF.close()
    
if __name__ == "__main__":
    auth = oauth(CONSUMER_KEY, CONSUMER_SECRET)
    delete_all(auth)
```
